Beautiful Soup

```python
used = "https://www.detik.com/terpopuler"
```

```python
from bs4 import BeautifulSoup
```

```python
url = used
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
list_content = soup.select('article')
```

```python
list_content
```

```python
len(list_content)
```

```python
list_content
```

```python
result = []
for i in list_content:
  data = dict()

  data['title'] = i.select_one('.media__title').text
  data['url'] = i.select_one('.media__title a').get('href')
  result.append(data)
```

```python
newframe = pd.DataFrame(result)
```
